```python
import numpy as np
import matplotlib.pyplot as plt
def plot_function(choice):
    x = np.linspace(-2*np.pi, 2*np.pi, 1000)
    if choice == 'sin':
        y = np.sin(x)
    elif choice == 'cos':
        y = np.cos(x)
    elif choice == 'tan':
        y = np.tan(x)
    else:
        print("Please enter 'sin', 'cos', or 'tan'.")
        return

    plt.plot(x, y)
    plt.title(f'Graph of {choice}(x)')
    plt.xlabel('x')
    plt.ylabel(f'{choice}(x)')
    plt.grid(True)
    plt.show()
choice = input("choose (sin, cos, or tan): ")
plot_function(choice.strip().lower(),)
```
